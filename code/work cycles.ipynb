```python
import seaborn as sns
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
schedule = pd.read_csv("DATA/observation data/observation_data.csv")
fuel_data = pd.read_csv("DATA/fuel_data_08_05.csv")
```

```python
schedule["timestamp"] = pd.to_datetime(schedule.timestamp).dt.strftime('%Y-%m-%d %H:%M')
fuel_data["timestamp"] = pd.to_datetime(fuel_data.timestamp, format="%Y-%m-%d %H:%M:%S").dt.strftime('%Y-%m-%d %H:%M')
```
